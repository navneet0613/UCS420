```python
import numpy as np
import matplotlib.pyplot as plt

# Q1: Matrix operations
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print("Sum of all elements:", np.sum(gfg))
print("Row-wise sum:", np.sum(gfg, axis=1))
print("Column-wise sum:", np.sum(gfg, axis=0))
```

```python
# Q2: Array sorting and filtering
array1 = np.array([10, 52, 62, 16, 16, 54, 453])
print("Sorted array:", np.sort(array1))
print("Indices of sorted array:", np.argsort(array1))
print("4 smallest elements:", np.sort(array1)[:4])
print("5 largest elements:", np.sort(array1)[-5:])

array2 = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
print("Integer elements:", array2[array2 == array2.astype(int)])
print("Float elements:", array2[array2 != array2.astype(int)])
```

```python
# Q3: Weekly Sales Dataset with Broadcasting
X = sum(ord(c) for c in "AB")  # Replace "AB" with your initials
sales = np.array([X, X+50, X+100, X+150, X+200])
tax_rate = ((X % 5) + 5) / 100
sales_after_tax = sales * (1 - tax_rate)
discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)

weeks = np.vstack([sales] * 3)
weeks = weeks * np.array([1, 1.02, 1.04])[:, None]
print("Weekly Sales after adjustments:\n", weeks)
```

```python
# Q4: Linspace and Function Evaluations
x = np.linspace(-10, 10, 100)
y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)

# Plot y = x^2
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='Y = x^2')
plt.title('Plot of Y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

# Plot y = sin(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y2, label='Y = sin(x)')
plt.title('Plot of Y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

# Plot y = e^x
plt.figure(figsize=(10, 6))
plt.plot(x, y3, label='Y = e^x')
plt.title('Plot of Y = e^x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

# Plot y = log(|x| + 1)
plt.figure(figsize=(10, 6))
plt.plot(x, y4, label='Y = log(|x| + 1)')
plt.title('Plot of Y = log(|x| + 1)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
