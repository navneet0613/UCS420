Q1. Ask the user to enter a value (e.g., M) for a mathematical function.

Generate x values from -10 to 10 using np.linspace().
Compute y values for:

o y= M⋅x 2

o y=M⋅sin(x)

Plot both functions in a single figure:
o Use different colors and line styles.
o Add legend, grid, and title.

```python
import matplotlib.pyplot as plt
import numpy as np

M = int(input("Enter a value for M: ")) #user input
x = np.linspace(-10, 10, 100)
y1 = M * x**2
y2 = M * np.sin(x)

plt.plot(x, y1, label=f'y = {M}x^2', color='blue', linestyle='-',linewidth=2)
plt.plot(x, y2, label=f'y = {M}sin(x)', color='red', linestyle='--',linewidth=2)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of function y = M*x^2 and y = M*sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```

Q.2 Create a dataset of five subjects and scores.
Convert it into a Pandas DataFrame.
Plot the scores using a Seaborn bar plot with:

o Different colors for each bar.

o Annotations on top of each bar.

o Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns

data = {
    'Subject': ['DSA', 'Probability', 'EDP', 'History', 'Art'],
    'Score': [88, 92, 79, 85, 90]
}

df = pd.DataFrame(data)
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
bar_plot = sns.barplot(x='Subject', y='Score', data=df, palette="Set2")
for p in bar_plot.patches:
    bar_plot.annotate(f'{p.get_height()}',
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center',
                      fontsize=12, color='black',
                      xytext=(0, 5), textcoords='offset points')
plt.title('Scores by Subject', fontsize=16)
plt.xlabel('Subject', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.grid(True)
plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().

Create a 2x2 subplot layout displaying:

o Line plot showing cumulative sum.

o Scatter plot with random noise.

Customize with titles, axis labels, and grids.

```python
r_no = 102497009
np.random.seed(r_no)
data = np.random.randn(50)
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].plot(np.cumsum(data), label='Cumulative Sum', color='blue')
axs[0].set_title('Cumulative Sum of Generated Values')
axs[0].set_xlabel('Index')
axs[0].set_ylabel('Cumulative Sum')
axs[0].grid(True)

# Plot 2: Scatter plot with random noise
axs[1].scatter(np.arange(len(data)), data + np.random.randn(50) * 0.5, label='Random Noise', color='red')
axs[1].set_title('Scatter Plot with Random Noise')
axs[1].set_xlabel('Index')
axs[1].set_ylabel('Values with Noise')
axs[1].grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

Q.4 Download Data-set from the below link
https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv
Apply ‘seaborn’ library to do the following.

1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
data = pd.read_csv(url)
print(data)

sns.set(style="whitegrid")

# Line plot for Total Profit of all months
plt.figure(figsize=(10, 6))
sns.lineplot(x='month_number', y='total_profit', data=data, marker='o', color='blue')
plt.title('\nTotal Profit Over Months', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Profit', fontsize=12)
plt.xticks(rotation=0)
plt.grid(True)
plt.show()

# Multiline plot for all product sales data
plt.figure(figsize=(12, 8))
sns.lineplot(x='month_number', y='facecream', data=data, marker='o', label='Product A Sales')
sns.lineplot(x='month_number', y='facewash', data=data, marker='o', label='Product B Sales')
sns.lineplot(x='month_number', y='toothpaste', data=data, marker='o', label='Product C Sales')
sns.lineplot(x='month_number', y='bathingsoap', data=data, marker='o', label='Product D Sales')

plt.title('\nSales Data of Different Products', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.xticks(rotation=0)
plt.legend()
plt.grid(True)
plt.show()

# Bar chart for all features/attributes
data.plot(kind='bar', figsize=(14, 8))
plt.title('\nBar Chart for All Features', fontsize=16)
plt.xlabel('Months', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```
